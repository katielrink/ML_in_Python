To ensure our data maintains high quality, so our results can be as accurate as possible, we need to clean our data. </br>

After an attempt to impute the null or unusable values, we found that it ended up skewing the values. Additionally the sheer amount of values caused overfitting in some of our calculations. Resultantly, we chose to simply remove the unusable data .</br>

First, we are going to remove columns that are not useful to our analysis. We can see that information such as Customer_ID, Month, Name, and SSN are general information that is extremely unlikely to have any affect on the trends we are analyzing. As a result, we can remove the columns to narrow our dataset. 

```python
#Print column names
data.columns
```

```python
#Remove Customer_ID, Month, Name, SSN
data.drop(['ID','Customer_ID', 'Month', 'Name', 'SSN'], axis=1, inplace=True)
```

Secondly, we are going to fill null values in the columns missing data. We also want to remove any illegal values

```python
#Replace Invalid Values
data =data.replace(r'[^\w\s.]|_', '', regex=True) 
#Replace all blank strings will null to be dropped
data.replace('', np.nan, inplace=True)
#Remove all rows with null values
data.dropna(inplace=True);
```

Thirdly, we want to clear out all duplicate data so our frequency analysis remains accurate. 

```python
#Find duplicate instances
duplicates = data[data.duplicated()]

#Remove all duplicates
data = data.drop_duplicates()
```

Fourthly, we would like to remove outliers from out dataset, so that our data analysis isn't skewed.

```python
#Determine outliers
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
#Remove outliers
data = data[~((data < (Q1 - 1.5*IQR)) | (data > (Q3 + 1.5*IQR))).any(axis = 1)]
data = data[(data['Num_Bank_Accounts'] >= 0)]
```

Finally, to make the data simpler to work with, we will replace credit score categories with values 1-3 for Good, Standard, or Poor. Additionally, we should replace the Credit_History_Age to numeric.

```python
#Replace Credot_History_Age to numeric
data['Credit_History_Age'] = data['Credit_History_Age'].str[:2]
data[['Credit_History_Age']] = data[['Credit_History_Age']].apply(pd.to_numeric)
```

Now that the data has been cleaned and simplified. We can describe it again to note a less skewed final view.

#### Types of Loans Being Requested


Another important understanding, is how the loans being requested tend to be distrubuted. To do this, I created a bar graph showing the number of each type of loan. For the most part the quantity of loans tend to be evenly distributed.

```python
#Create a dictionary to store the counts of all the various loan types
#Rerenced from Rashid Majeed at https://www.kaggle.com/code/rashidmajeedvirk/data-cleaning-and-visualization?kernelSessionId=102659541
loan_type_dict = dict()
loan_type_dict['Not Specified'] = 0
for value in data.Type_of_Loan:
    if 'Not Specified' in value:
        loan_type_dict['Not Specified'] += 1
        value = value.replace('Not Specified', '')
    values = re.split('Loan|and', value)
    for each_value in values:
        loan_type = each_value.strip(' ')
        if loan_type != '' : 
            if loan_type in loan_type_dict:
                loan_type_dict[loan_type] += 1
            else:
                loan_type_dict[loan_type] = 1
                
sns.set(rc = {'figure.figsize': (24, 12)})
sns.barplot(list(loan_type_dict.keys()), list(loan_type_dict.values()));
```

#### Trends Good Credit Scores and Occupation
One aspect that we wanted to inspect, is if a person's occupation would at all affect the Credit Score. Based on our below correlation computations, we can see that the correlation is incredibly low, therefore a person's credit score is not directly related to their occupation.From the bar graph, we can see the values tend to remain similar regardless of Occupation.

```python
#Calculate the Cross tabulation between Occupation and Credit Score
CrosstabResult = pd.crosstab(index=data['Occupation'],columns=data['Credit_Score'])
#Graph Correlation between information
CrosstabResult.plot.bar()
plt.show()
```

```python
# Performing Chi-sq test
ChiSqResult = chi2_contingency(CrosstabResult)
 
# P-Value is the Probability of H0 being True
# If P-Value&gt;0.05 then only we Accept the assumption(H0)
print('The P-Value of the ChiSq Test is:', ChiSqResult[1])
```

### Dimensionality Reduction

Dimensionality Reduction is simply the transformation of data from a higher-dimensional space to a lower-dimensional space, while trying to preserve the important properties of the dataset. The purpose of this process is to avoid working in high-dimensional spaces where data can be so sparse and analysis can get computationally tough. Uniform Manifold Approximation and Projection (UMAP) is a dimension reduction technique that can be used for visualization as well as general non-linear dimension reduction. It is built theoretically on advanced mathematics that is used to construct a high dimensional graph representation of the data then optimizes a low-dimensional graph to be as structurally similar as possible, or the so called "equivalent fuzzy topological structure". We try to follow the given link: https://github.com/lmcinnes/umap to implement dimensionality reduction using UMAP, then visualize and interpret the results.

```python
#Importing the umap library and constructing a UMAP Object
import umap.umap_ as umap
from sklearn.preprocessing import StandardScaler
reducer = umap.UMAP()

#n_neighbors=5,
#        min_dist=0.5,
#        n_components=2,
#        metric="euclidean")
#Remember 
print("The current shape of our dataset:",data.shape)

numerical_features = ["Annual_Income","Monthly_Inhand_Salary","Monthly_Inhand_Salary","Num_Bank_Accounts","Num_Credit_Card","Interest_Rate","Num_of_Loan","Delay_from_due_date","Num_of_Delayed_Payment","Changed_Credit_Limit","Num_Credit_Inquiries","Outstanding_Debt","Credit_Utilization_Ratio","Credit_History_Age","Total_EMI_per_month"]

credit_data = data[numerical_features].values 

print("The current shape of our dataset considering numerical features only:",credit_data.shape)

scaled_credit_data = StandardScaler().fit_transform(credit_data) #convert each feature into z-scores (number of standard deviations from the mean) for comparability.

embedding = reducer.fit_transform(scaled_credit_data)
print("The shape of our dataset after dimensionality reduction:",embedding.shape)

plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in data.Credit_Score.map({3:int(0),2:int(1),1:int(2)})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Credit Score dataset', fontsize=24)
plt.show()
```
